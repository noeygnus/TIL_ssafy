### Concept Check: INT8 vs INT4 양자화

실습에서는 FP16(16비트)과 INT4(4비트)를 비교했습니다. 이번에는 **INT8(8비트)과 INT4(4비트)**를 비교합니다.

**정밀도별 특성:**

| 정밀도 | 비트 수 | 메모리 (1B 파라미터) | 정밀도 손실 | 속도 |
|--------|---------|---------------------|------------|------|
| FP16 | 16비트 | 2GB | 없음 | 기준 |
| **INT8** | 8비트 | **1GB** | **낮음** | 빠름 |

**INT8 vs INT4 트레이드오프:**
- **INT8**: 메모리 50% 절감, 품질 손실 적음
- **INT4**: 메모리 75% 절감, 품질 손실 있음

### Guided Build: 추론 및 측정 함수 정의

실습에서 사용한 `measure_inference` 함수를 그대로 사용합니다.

````python
def measure_inference(model, tokenizer, prompt, max_new_tokens=128):
    """모델 추론을 수행하고 latency와 memory를 측정합니다."""
    
    # 메모리 측정 (추론 전)
    torch.cuda.synchronize()
    torch.cuda.reset_peak_memory_stats()
    
    # 입력 토큰화
    messages = [{"role": "user", "content": prompt}]
    inputs = tokenizer.apply_chat_template(
        messages,
        return_tensors="pt",
        return_dict=True,
        add_generation_prompt=True
    )
    
    # 입력을 모델 device로 이동
    input_ids = inputs["input_ids"].to(model.device)
    attention_mask = inputs["attention_mask"].to(model.device)
    
    input_length = input_ids.shape[1]
    
    # 추론 시간 측정
    start_time = time.time()
    # 재현성을 위한 seed 설정
    torch.manual_seed(42)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(42)
    with torch.no_grad():
        outputs = model.generate(
            input_ids,
            attention_mask=attention_mask,
            max_new_tokens=max_new_tokens,
            do_sample=True,
            temperature=0.7,
        )
    torch.cuda.synchronize()
    latency = time.time() - start_time
    
    # 메모리 측정
    memory = torch.cuda.max_memory_allocated() / 1024**3
    
    # 응답 디코딩
    response = tokenizer.decode(outputs[0][input_length:], skip_special_tokens=True)
    
    return response, latency, memory

print("measure_inference 함수 정의 완료")
````

### Guided Build: INT4 모델 로딩 (실습 복습)

먼저 INT4 모델을 로딩하여 기준 성능을 측정합니다. 이것은 실습에서 배운 내용의 복습입니다.

````python
# INT4 양자화 설정 (실습에서 사용한 것과 동일)
int4_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True,
)

# INT4 모델 로딩
print("INT4 모델 로딩 중...")
torch.cuda.reset_peak_memory_stats()
model_int4 = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=int4_config,
    device_map="auto",
)
tokenizer = AutoTokenizer.from_pretrained(model_name)

# 모델 메모리 측정
model_memory_int4 = torch.cuda.memory_allocated() / 1024**3
print(f"INT4 모델 메모리: {model_memory_int4:.2f} GB")
````

````python
# INT4 추론 테스트
test_prompt = "서울에서 부산까지 KTX로 얼마나 걸리나요?"
response_int4, latency_int4, memory_int4 = measure_inference(model_int4, tokenizer, test_prompt)

print("=" * 50)
print("[INT4 모델 추론 결과]")
print("=" * 50)
print(f"입력: {test_prompt}")
print(f"응답: {response_int4}")
print(f"\nLatency: {latency_int4:.2f}초")
print(f"피크 메모리: {memory_int4:.2f} GB")

# INT4 결과 저장
int4_results = {
    "model_memory": model_memory_int4,
    "latency": latency_int4,
    "peak_memory": memory_int4,
    "response": response_int4
}

# INT4 모델 메모리 해제
del model_int4
torch.cuda.empty_cache()
print("\nINT4 모델 메모리 해제 완료")
````

# Step 2: 양자화 설정 비교

BitsAndBytesConfig의 `bnb_4bit_use_double_quant` 옵션이 메모리에 미치는 영향을 확인합니다.

### Concept Check: double_quant (이중 양자화)

**double_quant란?**
- 양자화 상수(scale, zero_point)를 한번 더 양자화하여 추가 메모리 절감
- `bnb_4bit_use_double_quant=True`로 활성화

**작동 원리:**
```
일반 양자화:
  가중치(FP16) → 양자화된 가중치(INT4) + scale(FP32) + zero_point(FP32)
                                           ↑ 메모리 사용

이중 양자화:
  가중치(FP16) → 양자화된 가중치(INT4) + 양자화된 scale(INT8) + ...
                                           ↑ 추가 절감
```

**예상 효과:**
- 약 0.4bit/param 추가 절감 (이론치)
- 품질 손실은 거의 없음

### TODO 3: double_quant 설정 비교 (True vs False)

- **입력**: double_quant=False인 BitsAndBytesConfig
- **출력**: double_quant True/False 메모리 비교
- **힌트**: 앞서 로딩한 int4_results는 double_quant=True 상태

````python
# TODO: double_quant 설정 비교 (True vs False)

# INT8 모델 메모리 해제
del model_int8
torch.cuda.empty_cache()

# double_quant=False인 INT4 설정
int4_no_double_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=False,  # 이중 양자화 비활성화
)

# 모델 로딩
print("INT4 (double_quant=False) 모델 로딩 중...")
torch.cuda.reset_peak_memory_stats()
model_int4_no_double = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=int4_no_double_config,
    device_map="auto",
)

# 모델 메모리 측정
model_memory_no_double = torch.cuda.memory_allocated() / 1024**3
print(f"INT4 (double_quant=False) 모델 메모리: {model_memory_no_double:.2f} GB")
````

````python
# double_quant True vs False 비교
print("\n" + "=" * 60)
print("[double_quant 설정 비교]")
print("=" * 60)
print(f"{'설정':<25} {'모델 메모리 (GB)':>20}")
print("-" * 60)
print(f"{'double_quant=True':<25} {int4_results['model_memory']:>20.2f}")
print(f"{'double_quant=False':<25} {model_memory_no_double:>20.2f}")

double_quant_saving = model_memory_no_double - int4_results['model_memory']
double_quant_percent = (double_quant_saving / model_memory_no_double) * 100

print(f"\n{'절감량':<25} {double_quant_saving:>20.4f} GB ({double_quant_percent:.2f}%)")

print("\n✅ 분석:")
if double_quant_saving > 0:
    print(f"   - double_quant=True가 {double_quant_saving:.4f} GB 더 효율적")
    print(f"   - 이중 양자화로 양자화 상수의 메모리도 절감됨")
else:
    print(f"   - 이 모델 크기에서는 double_quant 효과가 미미함")
print(f"   - 대형 모델(7B+)에서는 효과가 더 뚜렷함")

# 메모리 해제
del model_int4_no_double
torch.cuda.empty_cache()
print("\n모델 메모리 해제 완료")
````

# Step 3: TTP 템플릿 설계

실습에서는 교통 도메인용 TTP 템플릿을 사용했습니다. 이번에는 **새로운 도메인용 TTP 템플릿을 직접 설계**합니다.

### Concept Check: TTP 템플릿 설계 원칙

**TTP(Test-Time Prompting)**란 추론 시점에 프롬프트를 강화하여 모델 출력의 품질을 안정화하는 기법입니다. 양자화로 인한 품질 저하를 재학습 없이 보완할 수 있습니다.

실습에서 배운 TTP 설계 원칙을 복습합니다:

1. **규칙을 명시하라**: 모델이 따라야 할 규칙을 명확하게
2. **출력 형식을 지정하라**: 원하는 응답 형식을 구체적으로
3. **Few-shot 예시를 포함하라**: 1-2개의 예시로 패턴 학습
4. **맥락을 제공하라**: 질문의 배경 정보 제공

**TTP-A vs TTP-B 비교:**

| 구분 | TTP-A (규칙 강화) | TTP-B (Few-shot) |
|------|------------------|-----------------|
| **방식** | 명시적 규칙·출력 형식 지정 | 입출력 예시로 패턴 학습 |
| **장점** | 제약이 명확, 구현 간단 | 복잡한 형식도 직관적 전달 |
| **단점** | 유연성 부족, 규칙 누락 가능 | 예시 작성 비용, 토큰 소비 증가 |
| **적용 시점** | 출력이 정형화된 경우 (숫자, 분류, 규격화된 형식) | 도메인 특화·복잡한 패턴이 필요한 경우 |

**새 도메인 선택:**
- 실습: 교통 도메인 (KTX 소요 시간 등)
- 과제: **의료 상담 도메인** 또는 **교육 Q&A 도메인** 선택

### TODO 4: 새로운 도메인용 TTP 템플릿 직접 작성

- **요구사항**: 의료 상담 도메인에 맞는 TTP 템플릿 2개(TTP-A, TTP-B)를 직접 설계하세요.
- **입력**: 의료 도메인 환경 변화 입력 샘플
- **출력**: MEDICAL_TTP_A (규칙 강화), MEDICAL_TTP_B (few-shot 포함)
- **힌트**: 
  - TTP-A: 출력 형식, 규칙, 주의사항 명시
  - TTP-B: few-shot 예시 포함
  - `{question}` 플레이스홀더 필수 포함

````python
# TODO: 새로운 도메인용 TTP 템플릿 직접 작성

# TTP-A: 규칙/형식 강화 템플릿
MEDICAL_TTP_A = """당신은 의료 상담 도우미입니다. 다음 질문에 답해주세요.

**규칙:**
1. 오타나 불명확한 표현이 있어도 환자의 의도를 파악하세요.
2. 증상이 모호하면 일반적인 조언을 제공하세요.
3. 반드시 "증상 → 가능한 원인 → 권장 조치" 순서로 답변하세요.
4. 심각한 증상은 반드시 병원 방문을 권유하세요.
5. 간결하고 이해하기 쉽게 답변하세요.

**질문:** {question}

**답변:**"""

# TTP-B: Few-shot 예시 포함 템플릿
MEDICAL_TTP_B = """당신은 의료 상담 도우미입니다. 아래 예시를 참고하여 질문에 답해주세요.

**예시:**
Q: 목이 아프고 열이 나요.
A: 증상으로 보아 감기 또는 편도염이 의심됩니다. 충분한 휴식과 수분 섭취를 권장하며, 38도 이상의 고열이 지속되면 병원 방문을 권유합니다.

Q: 배가 아파요 뭐가 문제일까요
A: 복통은 다양한 원인이 있을 수 있습니다. 소화불량, 위염 등이 일반적인 원인입니다. 식사 후 통증이 심해지거나 24시간 이상 지속되면 병원 진료를 받으시기 바랍니다.

이제 아래 질문에 동일한 형식으로 답해주세요.

**질문:** {question}

**답변:**"""

print("의료 도메인 TTP 템플릿 정의 완료")
print("- MEDICAL_TTP_A: 규칙/형식 강화")
print("- MEDICAL_TTP_B: Few-shot 예시 포함")
````

### TODO 5: 작성한 TTP로 환경 변화 테스트

- **요구사항**: 작성한 TTP 템플릿을 환경 변화 입력에 적용하고 효과를 비교하세요.
- **입력**: medical_samples 4개 × (TTP 없음, TTP-A, TTP-B) = 12회 추론
- **출력**: 입력 유형별 TTP 효과 비교
- **힌트**: template.format(question=prompt)로 템플릿 적용

````python
# TODO: 작성한 TTP로 환경 변화 테스트

def apply_ttp(template, question):
    """TTP 템플릿에 질문을 삽입합니다."""
    return template.format(question=question)

# 테스트 결과 저장
ttp_test_results = []

print("=" * 80)
print("[의료 도메인 TTP 테스트]")
print("=" * 80)

for sample in medical_samples:
    original_prompt = sample["prompt"]
    
    # TTP 없이 테스트
    response_no_ttp, _, _ = measure_inference(model_int4, tokenizer, original_prompt, max_new_tokens=150)
    
    # TTP-A 적용
    prompt_ttp_a = apply_ttp(MEDICAL_TTP_A, original_prompt)
    response_ttp_a, _, _ = measure_inference(model_int4, tokenizer, prompt_ttp_a, max_new_tokens=150)
    
    # TTP-B 적용
    prompt_ttp_b = apply_ttp(MEDICAL_TTP_B, original_prompt)
    response_ttp_b, _, _ = measure_inference(model_int4, tokenizer, prompt_ttp_b, max_new_tokens=150)
    
    # 결과 저장
    ttp_test_results.append({
        "type": sample["type"],
        "prompt": original_prompt,
        "no_ttp": response_no_ttp,
        "ttp_a": response_ttp_a,
        "ttp_b": response_ttp_b,
    })
    
    # 결과 출력
    print(f"\n[{sample['type']}] {original_prompt[:50]}...")
    print(f"  TTP 없음: {response_no_ttp[:100]}...")
    print(f"  TTP-A: {response_ttp_a[:100]}...")
    print(f"  TTP-B: {response_ttp_b[:100]}...")
    print("-" * 80)
````

````python
# TTP 효과 요약 (수동 품질 평가 기준 제시)
print("\n" + "=" * 60)
print("[TTP 효과 평가 기준]")
print("=" * 60)
print("""
품질 평가 기준 (1-5점):
- 5점: 정확하고 구조화된 답변, 적절한 조언 포함
- 4점: 대체로 정확, 일부 구조 부족
- 3점: 기본적인 답변, 구체성 부족
- 2점: 질문 이해 부족, 부적절한 답변
- 1점: 완전히 엉뚱한 답변

위 결과를 직접 확인하고 아래 표를 채워보세요:
""")

print(f"{'입력 유형':<12} {'TTP 없음':>12} {'TTP-A':>12} {'TTP-B':>12}")
print("-" * 60)
for result in ttp_test_results:
    # 실제로는 직접 평가해야 하지만, 예시로 채움
    print(f"{result['type']:<12} {'(평가필요)':>12} {'(평가필요)':>12} {'(평가필요)':>12}")

print("\n💡 위 응답들을 직접 읽고 품질을 평가해보세요.")
print("   TTP가 적용되었을 때 답변이 더 구조화되고 정확해지는지 확인하세요.")
````
