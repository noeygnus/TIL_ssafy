- 아래 코드는 Qwen3 모델을 다운로드 및 로드하는 코드입니다.
  - 한 1분 지나면 다운로드가 시작됩니다.
  - 다운로드 용량은 약 1.5G 입니다.

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

# AutoModelForCausalLM은 이전 단어들을 보고 다음 단어를 예측하는 모델을 불러올때 사용하는 클래스 입니다.
# 예시> 입력: 오늘 날씨가
#       모델이 예측: 좋다, 흐리다, 맑다
# GPT나 QWEN Llama 사용시 자주 사용하는 클래스 입니다.

# AutoModel - 일반 모델 (출력만, 보통 임베딩 모델)
# AutoModelForCausalLM (GPT와 같은 문장생성 출력형 모델)
# AutoModelForseq2seqLM (번역,요약에 특화된 모델)


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("사용 장치 : ", device)


model_name = "Qwen/Qwen3-0.6B"
print("모델 이름 : ", model_name)

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto", # 모델에서 사용하는 데이터 타입 ( flaot32 32비트 부동소숫점 쓸것인지 fp64를 쓸것인지 모델이 최적화된 타입으로 자동설정 )
    device_map="auto"   # VRAM 부족해서 GPU 메모리 꽉 차면 알아서 일부를 CPU RAM으로 올려서 거기서 연산
)

````

- 아래 코드는 추론하는 코드입니다.
  - GPU 수행시 : 1회 추론시 약 3초 소요
  - CPU 수행시 : 1회 추론시 약 12초 소요

````python
prompt = "너의 이름이 뭐니?"

messages = [
    {
        "role": "user",
        "content": prompt
    }
]

# 토큰화
text = tokenizer.apply_chat_template(
    messages,
    tokenize=False, # 단순 문자열로 반환, 만약에 True면 토큰을 id 리스트로 반환
    add_generation_prompt=True,  # 지금까지의 대화는 다 입력이고, eos 이제 어시스턴트의 답변을 생성할 차례라는 것을 알리는 시작 토큰 생성 (없으면 답변생성 안함)
    enable_thinking=False) # 답변생성시 생각 과정을 길게 쓸게 할지 정하는 옵션


# 모델에 입력하기 위해 PyTouch의 Tensor로 변경
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

# 모델 추론
# generated_ids: 출력결과는 인덱싱 이다.
# 모델이 토큰id를 입력으로 받으면, 입력받은 토큰id가지고 다음 토큰id를 예측 및 생성
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=500,
    temperature=0.7, # 얼마나 다양한 단어를 사용할지
    top_p=0.8,
)

# top_p : 확률 누적합이 p(예: 0.8)가 넘을 때!
# 딱 거기까지만 후보를 남기고, 나머지는 버림
# 예시
# 좋다(0.4), 맑다(0.3), 흐리다(0.2), 춥다(0.05),... 단어 후보군이 있으면
# 상위 세 단어(좋다, 맑다, 흐리다)만 남기고 나머지 버림!!!
# LLM이 다음에 나올 토큰(후보군)에 대한 점수(로짓점수)를 softmax를 이용해서 확률로 바꾼값이 0.4 0.3 값임

print(generated_ids)
print(generated_ids[0])

# 디코딩 (토큰 ID를 문자로 변환)
output_ids = generated_ids[0][len(model_inputs.input_ids[0]):]
content = tokenizer.decode(output_ids, skip_special_tokens=True)

print(content)

# 모든 LLM은 원래 토큰 ID를 출력하는 모델이고
# 텍스트는 우리가 디코딩해서 보는 것일 뿐입니다.
````

### 데이터 생성하기

- 프롬프트를 입력하여 Text를 생성해보겠습니다.

````python
prompt = """
원본 데이터를 기반으로 새로운 데이터들을 만들어줘
같은 의미를 가지고 있지만, 다양한 표현으로 바꾸어줘

원본 데이터 : "안녕, 난 철수야. KFC를 좋아하지"
만들고 싶은 데이터 예시는 다음과 같아
1. 안녕하세요, 철수입니다. 오늘은 KFC의 인기 비결에 대해 알아보겠습니다.
2. 안녕~, 철수예요. KFC의 로고 컬러는 언제 봐도 강렬하죠.
3. 여러분 안녕하세요. 저는 철수랍니다. 수업 후엔 KFC 치킨을 즐긴답니다.

새로운 문장 1개로만 대답해줘
"""

messages = [
    {
        "role": "user",
        "content": prompt
    }
]

# 토큰화
text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True, enable_thinking=False)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

# 모델 추론
generated_ids = model.generate(
    **model_inputs,
    max_new_tokens=500,
    temperature=0.7,  # 사용 단어의 창의성 농도
    top_p=0.8,        # 후보 단어를 어디까지 포함할지, 단어 선택 후보를 넓게 볼지 적게 볼지
)

# 디코딩 (토큰 ID를 문자로 변환)
output_ids = generated_ids[0][len(model_inputs.input_ids[0]):]
content = tokenizer.decode(output_ids, skip_special_tokens=True)

print(content)
````

- 출력되는 소스코드를 편집하면, 5000개 문장도 생성할 수 있을 것 같습니다.
  - for 문을 돌려서, 반복적으로 수행하면 되겠습니다.
  - 한 문장 생성당 Runtime 사용 1초씩 사용됩니다.
    - 5000초가 드니까, 코랩 Runtime 2시간이면 되겠네요 ^^
    - 코랩 대신, 좋은 GPU가 달린 추론 서버를 돌리면, 걱정은 없겠습니다.

- 아래 주석 코드를 바꿔가면서, 여러번 실행해보세요!

````python
prompt = """
원본 데이터를 기반으로 새로운 데이터들을 만들어줘
같은 의미를 가지고 있지만, 다양한 표현으로 바꾸어줘

원본 데이터 : "안녕, 난 철수야. KFC를 좋아하지"
만들고 싶은 데이터 예시는 다음과 같아
1. 안녕하세요, 철수입니다. 오늘은 KFC의 인기 비결에 대해 알아보겠습니다.
2. 안녕~, 철수예요. KFC의 로고 컬러는 언제 봐도 강렬하죠.
3. 여러분 안녕하세요. 저는 철수랍니다. 수업 후엔 KFC 치킨을 즐긴답니다.

새로운 문장 1개로만 대답해줘
"""

messages = [
    {
        "role": "user",
        "content": prompt
    }
]

# 토큰화
text = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True, enable_thinking=False)
model_inputs = tokenizer([text], return_tensors="pt").to(model.device)

# 10개 문장을 만들자!
for i in range(10):

  # 모델 추론
  generated_ids = model.generate(
      **model_inputs,
      max_new_tokens=500,

      # ==================================
      # 창의성 넘치지만, 생성된 데이터 품질이 낮음
      #temperature=1.0,
      #top_p=0.9,

      # 창의성 없지만, 생성된 데이터 품질이 높음
      #temperature=0.6,
      #top_p=0.5,

      # Qwen3 권장 튜닝값
      temperature=0.7,
      top_p=0.8,
      # ==================================
  )

  # 디코딩 (토큰 ID를 문자로 변환)
  output_ids = generated_ids[0][len(model_inputs.input_ids[0]):]
  content = tokenizer.decode(output_ids, skip_special_tokens=True)

  print(content)
````

- 다음과 같은 상황이 될때, 두 파라미터를 조율하시면 됩니다.
  - AI가 만든 데이터가 똑같은 값이 반복적일때 : 창의성을 높이기 위해 Temp, top_p 둘다 UP
  - AI가 만든 데이터가 말도 안되는 값을 자주 만들때 : 정확도를 높이기 위해 Temp, top_p 둘다 Down

### 부정확한 데이터 필터링하기
- Temperature와 Top_p를 적절히 조율 했지만, 부정확한 데이터가 간헐적으로 나옵니다.

- 이제 아래 두 코드를 실행해서 잘 동작되는지 확인해봅시다.
  - Upstage가 OpenAI API 형식을 그대로 지원합니다.
  - 그래서 Upstage 모델을 사용할 것이지만, OpenAI 라이브러리를 그대로 사용합니다.

````python
from google.colab import userdata
from openai import OpenAI

# Upstage로 연동합니다.
client = OpenAI(
    api_key=userdata.get('OPEN_AI'),
    base_url="https://api.upstage.ai/v1"
)
````

````python
# API: client.chat.completions.create
# 완전히 Open AI의 API와 동일하게 호출
# 차이점은 model을 solar-pro3라는 Upstage 모델을 사용하는 것뿐


response = client.chat.completions.create(
    model="solar-pro3", # 개발 및 테스트에는 저렴한 모델을 사용해주세요.
    messages=[
        {"role": "user", "content": "KFC 버거 추천좀 해줘, 두줄 요약"}
    ],
    temperature=0.7,
    max_tokens=200
  )

print(response)
print("-----")
print(response.choices[0].message.content)
````

- 준비가 다 되었습니다.

### LLM as a judge를 이용한 데이터 필터링
- 프롬프트를 이용하여 평가를 진행할 LLM(Solar-pro3)에게 필터링을 지시하면 끝입니다.
  - 실제 LLM as a judge도 같은 방식입니다.

- 다음과 같은 용어를 사용하겠습니다.
  - **Judge LLM** : 평가하는 LLM, Solar-pro3 모델이 되겠습니다.
  - **Target LLM** : 평가 받는 LLM, Qwen3-0.6B 모델이 되겠습니다.

- 코드로 표현하면 다음과 같습니다.
  - 영어로 프롬프트를 작성하면, 일반적으로 성능이 더 좋긴하지만 편의상 한글로 했습니다.

````python
prompt = """
자네는 LLM as a judge의 평가자 역할을 담당하네.
[Question] 은 Target LLM이 받은 질문이고, [Answers]는 Target LLM이 답변한 응답들이지.
다음 Target LLM의 응답들의 품질을 1 ~ 5점으로 평가했을 때,
최종 결과를 응답 순서대로 List를 만들어주거라.

[Question]
원본 데이터를 기반으로 새로운 데이터들을 만들어줘
- 같은 의미를 가지고 있지만, 다양한 표현으로 바꾸어줘
- 사람이 이해할 수 있는 문장이어야해
- 원본 데이터 : "안녕, 난 철수야. KFC를 좋아하지"

[Answers]
안녕하세요, 철수입니다. KFC을 사랑하지 않는일은 없습니다.
안녕하세요, 저는 철수입니다. 나이는 없었지만 치킨의 빈도가 흔하며, 이 곳에서 KFC를 즐기려고 했습니다.
안녕하세요, 저는 철수답니다. KFC의 로고에 대한 칭호는 언제 봐도 그 깎은색빛 덕수아니죠.
안녕, 철수입니다. KFC가 꼽히는 소리가 가고 있을 때마다 소통이 가는 날입니다.
안녕하세요. 저는 철수랍니다. KFC을 좋아하며, 치킨의 아침 열면 푸짐을 추천하세요!
안녕, 철수입니다. KFC를 정말 즐바요.
"안녕, 오늘 KFC로 만난 친구가 철수죠. 철수라는 표현으로 뒷도깨진 이야기로 보려주세요!"
안녕하세요, 철수입니다. 우리들의 KFC에 관심을 가지고 있습니다.

최종 답변은 다음 List 형태로 품질 점수와 Best 문장 3개와 Worst 문장 3개를 출력하거라.
예시
Result : [2, 3, 1, 5, 3, 4, 2...]
Best1 : ... (이유 한줄)
Best2 : ... (이유 한줄)
Best3 : ... (이유 한줄)
Worst1 : ... (이유 한줄)
Worst2 : ... (이유 한줄)
Worst3 : ... (이유 한줄)
"""

response = client.chat.completions.create(
    model="solar-pro3",
    messages=[
        {"role": "user", "content": prompt}
    ]
)

print(response.choices[0].message.content)
````

# 4. 데이터 생성 방법의 분류
- 명확히 데이터 증강과 합성 데이터를 구분해봅시다.
  - AI를 이용하여 데이터를 생성하는 방법은 대표적으로 두가지가 있습니다.
    - 데이터 증강 : 원본text의 의미는 그대로고 표현만 변경하는 것이지요.
    - 합성데이터 : 새로 창작된 text로 원본의 정도 + 원문에 없는 정보가 추가되는 경우




###  데이터 생성 방법 퀴즈!

### 사전준비 - 이미지 다운로드 후 출력하는 코드 이해
- 먼저 데이터를 증강하기 전에, 이미지 다운받고 출력하는 코드를 이해해봅시다.
  1. 이미지를 requests 라는 Library로 다운로드 받습니다.
  2. BytesIO로 바이트 값으로 변경 후 이미지 타입으로 바꿉니다.
  3. 이후 Matplot으로 출력해봅니다.

````python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

# 이미지 경로
url = "https://cdn.pixabay.com/photo/2016/12/17/00/06/bowser-1912623_1280.jpg"

# 다운로드 (request로 데이터 다운받고, 이미지 타입으로 변환)
response = requests.get(url)
img_origin = Image.open(BytesIO(response.content)).convert("RGB")
# 픽사베이 이미지 파일을 로컬에 저장하지 않고 가상 메모리 위에 저장
# 코랩 가상컴퓨터에 파일로 저장하는게 아니라 메모리에 해당 이미지의 비트를 잠시 저장하는거지?


print(type(img_origin))


# 출력하기
plt.imshow(img_origin)
plt.axis("off")
plt.show()
````

### 다음은 이미지 데이터 증강을 해보겠습니다.
#### Torchvision 라이브 러리를 사용할 것입니다.
- 이미지 처리를 위해서 나온, PyTorch의 공식 Library 입니다.
- 이미지 AI를 연구하시는 분들이 많이 사용하는 Library 입니다.
  - Torchvision을 이용하면 간결한 코드로 이미지 전처리를 할 수 있습니다!
  - CNN 수업할 때 사용하는 그 Torchvision이 맞습니다.
  

<br/>


````python
from torchvision import transforms # torchvision안에 전처리 기능들이 모인 클래스
from torchvision.utils import save_image


# 이미지 전처리 어떻게 할지 정의 후,
base_transform = transforms.Compose([
    transforms.CenterCrop(640), # 640 x 640 이미지로 Crop (이미지 자르기), 이미지가 너무 크면 연산량만 늘어남
    transforms.Resize((224, 224)), # 모델 입력 크기에 맞게 (224 x 224) 사이즈로 위에서 자른 이미지를 축소
    transforms.ToTensor()
])

# 이전 코드셀에서 다운받은 이미지에 적용하기
base_img = base_transform(img_origin)


# 증강한 데이터에 대한 정의
aug_transform = transforms.Compose([
    transforms.CenterCrop(640), # 중앙 기준으로 이미지 정리 (불필요한 배경 제거)
    transforms.Resize((400, 400)), # 위아래좌우 랜덤하게 224 사이즈로 자르기 위해 이미지 크기 확대 (다양성확보)
    transforms.RandomHorizontalFlip(p=0.5),  # 50% 확률로 좌우반전(다양성확보)
    transforms.RandomRotation(degrees=10),   #  -10 ~ 10도 사이 랜덤 회전(다양성확보)
    transforms.RandomApply([transforms.RandomGrayscale(p=1.0)], p=0.2), # 20% 확률로 흑백사진(색상 의존도 감소)
    transforms.ColorJitter(brightness=0.3, contrast=0.3, saturation=0.3), # 밝기, 대조, 노출값 랜덤하게 변경(조명변화대응)
    transforms.RandomCrop((224, 224)), # 랜덤한 위치로 Crop
    transforms.ToTensor()
])

aug_imgs = []
for i in range(5):
  aug_img = aug_transform(img_origin)
  aug_imgs.append(aug_img) # 전처리된 이미지 5개 생성


# 시각화 (아래 코드는 이쁘게 출력하기 위한 코드입니다.)


to_pil = transforms.ToPILImage() #텐서를 필로우(PIL) 이미지로 바꿈

plt.figure()
plt.title('Original')
plt.imshow(to_pil(base_img))
plt.axis("off")

fig, axes = plt.subplots(1, 5, figsize=(7, 7))
for i, aug in enumerate(aug_imgs):
    axes[i].imshow(to_pil(aug))
    axes[i].set_title(f'arg{i+1}')
    axes[i].axis("off")

plt.show()


# 파일로 저장
save_image(base_img, "aug_grid.jpg")
for i, img in enumerate(aug_imgs):
    save_image(img, f"aug_image_{i+1}.jpg")
````
